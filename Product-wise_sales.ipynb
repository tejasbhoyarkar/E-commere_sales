# Task: Load product-Wise Dataset

```python
product_sales = df[["Order ID","Product Category" , "Product" , "Quantity Ordered", "Price Each"]]
product_sales.shape
```

```python
product_sales_clean = product_sales.drop_duplicates().reset_index(drop=True)
product_sales_clean.shape
```

```python
product_sales_clean.columns
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
```

# Maximum Sales Product (Overall)

```python
product_sales = df.groupby('Product')['Sales'].sum()
top_product = product_sales.idxmax()
top_product_sales = product_sales.max()
top_product, top_product_sales
```

# Peak Sales Hour

```python
hour_sales = df.groupby('Hour')['Sales'].sum()

peak_hour = hour_sales.idxmax()
peak_hour_sales = hour_sales.max()

peak_hour, peak_hour_sales
```

# Product-wise total sales


```python
product_sales = (df.groupby(['Product Category', 'Product'])['Sales'].sum().reset_index())

product_sales
```

# Find TOP (Maximum) Selling Products

```python
top_products = product_sales.sort_values('Sales', ascending=False)
top_products.head(10)
```

# Product-wise Quantity Sold

```python
product_quantity = (df.groupby('Product')['Quantity Ordered'].sum().reset_index().sort_values('Quantity Ordered', ascending=False))

product_quantity.head(10)
```

# Category-level Insight

```python
category_sales = (df.groupby('Product Category')['Sales'].sum().reset_index()
    .sort_values('Sales', ascending=False))
category_sales
```
