```python
import pandas as pd 
```

```python
original_df = pd.read_csv('Sales_Data.csv')
original_df.head()
```

```python
df = original_df.copy()
df = pd.read_csv('Sales_Data.csv')
print(df.shape)
print(df.columns)
```

# Monthly Sales Summary CSV

```python
monthly_sales_summary = (
    df.groupby('Month')['Sales']
    .sum()
    .reset_index()
)

monthly_sales_summary.to_csv("Monthly_Sales_Summary.csv", index=False)
```

# Product Sales Summary CSV

```python
product_sales_summary = (df.groupby(['Product Category', 'Product'])
    .agg(
    Total_Quantity=('Quantity Ordered', 'sum'),
        Total_Sales=('Sales', 'sum')
    )
    .reset_index()
)

product_sales_summary.to_csv("Product_Sales_Summary.csv", index=False)
```

# City-wise Sales Summary CSV

```python
city_sales_summary = (df.groupby('City')['Sales'].sum().reset_index()
)

city_sales_summary.to_csv("City_Sales_Summary.csv", index=False)
```
